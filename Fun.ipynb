```python
X_Cat,X_Cat1,X_Num,Y,order_list = Ch.Data_Selection(df)
```

```python
cat_fea = X_Cat.columns
cat_fea1 = X_Cat1.columns
num_fea = X_Num.columns
print(num_fea)
print(cat_fea)
print(cat_fea1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
num_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
     ('scaler', StandardScaler())
     ])
num_transform
```

```python
order_list
```

```python
cat_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder',OrdinalEncoder(categories=order_list,dtype=np.int64))
])
cat_transform
```

```python
cat_transform1 = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent'))
])
cat_transform1
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("num",num_transform,num_fea),
        ("cat",cat_transform,cat_fea),
        ("cat1",cat_transform1,cat_fea1)
    ]
)
preprocess
```

```python
pline = Pipeline(steps=[
    ('preprocess',preprocess)
])
pline
```

```python
x = pd.concat([X_Cat,X_Cat1,X_Num],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.25,random_state=42)
x_train
```
